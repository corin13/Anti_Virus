```python
from scapy.all import *

# Pcap   
pcap_file = 'captured_packets.pcap'

#     
packet_lengths = []

# Pcap  
packets = rdpcap(pcap_file)

#       
for i, pkt in enumerate(packets):
    packet_lengths.append(len(pkt))
    print(f"Packet {i+1}: Length = {len(pkt)} bytes")

#     
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(range(len(packet_lengths)), packet_lengths, marker='o', linestyle='-', color='b')
plt.xlabel('Packet Number')
plt.ylabel('Packet Length (bytes)')
plt.title('Packet Lengths in Pcap File')
plt.grid(True)
plt.tight_layout()
plt.show()
```
