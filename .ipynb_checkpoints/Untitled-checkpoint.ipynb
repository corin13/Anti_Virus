```python
import matplotlib.pyplot as plt

#    
def read_data_from_file(filename):
    with open(filename, 'r') as file:
        data = file.read()
    return data

# CPU   
def visualize_cpu_usage(cpu_data):
    try:
        #  :   
        lines = cpu_data.splitlines()
        start_line = lines.index('## CPU Usage:') + 1  #   
        cpu_data = lines[start_line:]  #    

        #  
        cpu_cores = []
        cpu_usage = []
        for line in cpu_data:
            if line.startswith('CPU'):
                parts = line.split()
                if len(parts) < 3:
                    continue
                core = parts[1]
                usage = float(parts[-1])
                cpu_cores.append(core)
                cpu_usage.append(usage)

        # 
        plt.figure(figsize=(10, 6))
        plt.bar(cpu_cores, cpu_usage, color='skyblue')
        plt.xlabel('CPU Core')
        plt.ylabel('Usage (%)')
        plt.title('CPU Usage')
        plt.ylim(0, 100)
        plt.grid(True)
        plt.tight_layout()
        plt.show()

    except ValueError as ve:
        print(f"ValueError occurred: {ve}")
    except IndexError as ie:
        print(f"IndexError occurred: {ie}")
    except Exception as e:
        print(f"Exception occurred: {e}")


#    
def visualize_network_usage(network_data):
    try:
        #  :   
        lines = network_data.splitlines()
        start_line = lines.index('## Network Usage:') + 1  #   
        network_data = lines[start_line:]  #    

        #     
        interfaces = []
        usages = []
        for line in network_data:
            parts = line.split()
            if len(parts) < 3:
                continue  #      
            
            iface = parts[0]
            try:
                usage = float(parts[-2])  #   '%'   float 
                interfaces.append(iface)
                usages.append(usage)
            except ValueError:
                continue

        # 
        plt.figure(figsize=(10, 6))
        plt.bar(interfaces, usages, color='orange')
        plt.xlabel('Network Interface')
        plt.ylabel('Usage (%)')
        plt.title('Network Usage')
        plt.ylim(0, 100)
        plt.grid(True)
        plt.xticks(rotation=45)  # x  45     
        plt.tight_layout()
        plt.show()

    except ValueError as ve:
        print(f"ValueError occurred: {ve}")
    except IndexError as ie:
        print(f"IndexError occurred: {ie}")
    except Exception as e:
        print(f"Exception occurred: {e}")

#    
def visualize_disk_usage(disk_data):
    try:
        #  :   
        lines = disk_data.splitlines()
        start_line = lines.index('## Disk Usage:') + 1  #   
        disk_data = lines[start_line:]  #    

        #  
        disks = []
        usages = []
        for line in disk_data:
            if line.startswith('/dev/'):
                parts = line.split()
                if len(parts) < 6:
                    continue
                disk = parts[0]
                try:
                    usage = float(parts[-2][:-1])  #   '%'   float 
                    disks.append(disk)
                    usages.append(usage)
                except ValueError:
                    continue

        # 
        plt.figure(figsize=(10, 6))
        plt.barh(disks, usages, color='lightgreen')
        plt.xlabel('Usage (%)')
        plt.ylabel('Disk')
        plt.title('Disk Usage')
        plt.xlim(0, 100)
        plt.grid(True)
        plt.tight_layout()
        plt.show()

    except ValueError as ve:
        print(f"ValueError occurred: {ve}")
    except IndexError as ie:
        print(f"IndexError occurred: {ie}")
    except Exception as e:
        print(f"Exception occurred: {e}")

#     
def visualize_all_resources(data):
    visualize_cpu_usage(data)
    visualize_network_usage(data)
    visualize_disk_usage(data)

#  :
if __name__ == '__main__':
    filename = 'usage_data.txt'  #    
    data = read_data_from_file(filename)

    visualize_all_resources(data)
```
